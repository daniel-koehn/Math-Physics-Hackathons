```python
# Import libraries
from matplotlib import pyplot
import random
import numpy
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

## Barnsley Fern

```python
# define parameters
itermax = 1100000
a0 = 1
a1 = 86
a2 = 93
```

```python
xs = numpy.zeros(itermax+1)
ys = numpy.zeros(itermax+1)

x = 0
y = 0
for n in range(itermax):
    
    r = random.random()  # pick random number
    r = r * 100
    
    xn = x
    yn = y
    
    if r < a0:  # stem
        x = 0
        y = 0.16 * yn
        
    elif r < a1: # successively smaller leaflets
        x = 0.85 * xn + 0.04 * yn
        y = -0.04 * xn + 0.85 * yn + 1.6
        
    elif r < a2: # largest left-hand leaflet 
        x = 0.20 * xn - 0.26 * yn
        y = 0.23 * xn + 0.22 * yn + 1.6
        
    else:        # largest right-hand leaflet 
        x = -0.15 * xn + 0.28 * yn
        y = 0.26 * xn + 0.24 * yn + 0.44
        
    xs[n] = x
    ys[n] = y
```

#### Plot Barnsleys Fern

```python
fig = pyplot.figure(figsize=(10., 10.))
pyplot.tight_layout()
pyplot.plot(xs,ys,'g,')

# export Barnsleys Fern as Jpeg 
pyplot.savefig('Barnsleys_Fern.jpeg', bbox_inches='tight', dpi=150)
pyplot.show()
```
