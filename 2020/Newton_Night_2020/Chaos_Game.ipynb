## Chaos game for a triangle

```python
# define parameters here
npoints = 3
npoints +=1

points_x = numpy.zeros(npoints)
points_z = numpy.zeros(npoints)

# define triangle
points_x = [0,1,0.5]
points_z = [0,0,1]

# define initial position
pos_init_x = 0.25 * numpy.ones(1)
pos_init_z = 0.25 * numpy.ones(1)

itermax = 400000
```

#### Let's run the chaos game for the triangle

```python
# allocate position array
pos_x = numpy.zeros(itermax+1)
pos_z = numpy.zeros(itermax+1)

# Set initial position
pos_x[0] = pos_init_x
pos_z[0] = pos_init_z

for n in range(itermax):
    
    # pick random corner point
    pick_corner = numpy.random.randint(npoints-1, size=1)
    
    # update position to midpoint between current position and random corner point
    pos_x[n]  = pos_x[n-1] + (points_x[pick_corner[0]] - pos_x[n-1]) / 2.
    pos_z[n]  = pos_z[n-1] + (points_z[pick_corner[0]] - pos_z[n-1]) / 2.
```

```python
# Plot result of chaos game
fig = pyplot.figure(figsize=(10., 10.))
pyplot.tight_layout()
pyplot.plot(points_x,points_z,'r*') # plot polygon corner points 
pyplot.plot(pos_x,pos_z,'b,')       # plot result of chaos game

# save chaos game result as pdf 
pyplot.savefig('chaos_game_triangle.jpeg', bbox_inches='tight', dpi=150)
pyplot.show()
```

## Restricted chaos game 1 for a square

```python
# define parameters here
npoints = 4
npoints +=1

points_x = numpy.zeros(npoints)
points_z = numpy.zeros(npoints)

# define square
points_x = [0,1,1,0]
points_z = [0,0,1,1]

# define initial position
pos_init_x = 0.25 * numpy.ones(1)
pos_init_z = 0.25 * numpy.ones(1)

itermax = 2000000
```

#### Let's run the restricted chaos game 1 for the square

```python
# allocate position array
pos_x = numpy.zeros(itermax+1)
pos_z = numpy.zeros(itermax+1)
picks = numpy.zeros(itermax+1)

# Set initial position
pos_x[0] = pos_init_x
pos_z[0] = pos_init_z

for n in range(itermax):
    
    # pick random corner point
    pick_corner = numpy.random.randint(npoints-1, size=1)
    
    # If two consecutively corners are equal ...
    if(n>2):
        if(picks[n-1] == picks[n-2]):
            
            # pick a corner not adjacent to picks[n-1] == picks[n-2]
            if(picks[n-1]==0):
                pick_corner[0] = 2
                
            if(picks[n-1]==1):
                pick_corner[0] = 3
                
            if(picks[n-1]==2):
                pick_corner[0] = 0
                
            if(picks[n-1]==3):
                pick_corner[0] = 1    
    
    # update position to midpoint between current position and random corner point
    pos_x[n]  = pos_x[n-1] + (points_x[pick_corner[0]] - pos_x[n-1]) / 2.
    pos_z[n]  = pos_z[n-1] + (points_z[pick_corner[0]] - pos_z[n-1]) / 2.
    
    # store current pick
    picks[n] = pick_corner[0]
```

```python
# Plot result of chaos game
fig = pyplot.figure(figsize=(10., 10.))
pyplot.tight_layout()
pyplot.plot(points_x,points_z,'r*') # plot polygon corner points 
pyplot.plot(pos_x,pos_z,'b,')       # plot result of chaos game

# save chaos game result as pdf 
pyplot.savefig('chaos_game_square_restr1.jpeg', bbox_inches='tight', dpi=150)
pyplot.show()
```
